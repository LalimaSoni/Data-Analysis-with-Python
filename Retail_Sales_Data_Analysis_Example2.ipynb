### Retail_Sales_Data_Analysis_Example2

This Data anaysis covers below mentioned questions.
Considering given data set is already cleaned.

1. How many unique shops are in the dataset?
<br>2. How many unique preferred and cross border shops are in the dataset?
<br>3. How many products have zero sold count?
<br>4. How many products were created in the year 2018?
<br>5. Top 3 Preferred shops’ shopid that have the largest number of unique products
<br>6. Top 3 Categories that have the largest number of unique cross-border products
<br>7. Find Top 3 shopid with the highest revenue (Assumption: the product price has not been changed.) 
<br>8. Find number of products that have more than 3 variations (do not include products with 3 or fewer variations)
<br>9. Identify duplicated listings within each shop (If listing A and B in shop S have the exactly same product title, product detailed description, and price, both listing A and B are considered as duplicated listings)
<br>10. Mark those duplicated listings with True otherwise False and store the marking result in a new column named “is_duplicated” 
<br>11. Find duplicate listings that has less than 2 sold count and store the result in a new excel file named “duplicated_listings.xlsx”
<br>12. Find the preferred shop shopid that have the most number of duplicated listings

```python
import pandas as pd
data = pd.read_excel("Test_Pandas.xlsx")
data.head()
```

```python
columns = data.columns
print(columns)
```

```python
#1 	How many unique shops are in the dataset?
print(data.shopid.nunique())
```

```python
#2 How many unique cross border shops are in the dataset?
print(data.cb_option.nunique())
```

```python
#2 How many unique preferred shops are in the dataset?
print(data.is_preferred.nunique())
```

```python
#3 How many products have zero sold count?
print(len(data[data["sold_count"]==0]))
```

```python
#4 How many products were created in the year 2018?
data.info()
```

```python
data["item_creation_date"] = data["item_creation_date"].astype(str)
data["year"] = data["item_creation_date"].apply(lambda x : x.split ("-")[0])
data.head()
```

```python
#4 .	How many products were created in the year 2018?
print(len(data[data["year"] == '2018']))
```

```python
#5 	Top 3 Preferred shops’ shopid that have the largest number of unique products
#data["itemid"].nunique()
s1 = data[data["is_preferred"] == 1]
s2 = s1.groupby("shopid", as_index=False).count()

s3 = s2.nlargest(3,"itemid")
s3[["shopid","itemid"]]
```

```python
#6	Top 3 Categories that have the largest number of unique cross-border products

s4 = data[data["cb_option"] == 1]
s5 = s4.groupby("category", as_index = False).count()
s6 = s5.nlargest(3,"itemid")
s6[["category","itemid"]]
```

```python
 #7 	Find Top 3 shopid with the highest revenue (Assumption: the product price has not been changed.) 
    
data["Sales"] = data["price"] * data["sold_count"]
revenue = data.groupby("shopid", as_index = False).sum()
s7 = revenue.nlargest(3, "Sales")
s7[["shopid", "Sales"]]
```

```python
## 12 Find the preferred shop shopid that have the most number of duplicated listings
d7 = data[data["is_preferred"] == 1]
d8 = d7.groupby("shopid", as_index = False).count()
d9 = d8.nlargest(1,"is_duplicated")
```
